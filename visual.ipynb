```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = ["Attr. Bind.", "Obj. Rel.", "Num.", "Complex"]
baseline = [0.3173, 0.3110, 0.2375, 0.3082]
config1 = [0.3185, 0.3128,  0.2713, 0.3094]
plt.bar(categories, baseline-0.2)
plt.bar(categories, config1+0.2)
plt.ylim(0.2, 0.34)
```

```python
x = np.arange(4)
baseline = [0.3173, 0.3110, 0.2375, 0.3082]
config1 = [0.3185, 0.3128,  0.2713, 0.3094]
width = 0.4

plt.bar(x-0.2, baseline, width)
plt.bar(x+0.2, config1, width)
plt.xticks(x, ["Attr. Bind.", "Obj. Rel.", "Num.", "Complex"])
plt.xlabel("Metrics")
plt.ylabel("Values")
plt.legend(["Baseline", "Config 1"])
plt.ylim(0.2, 0.34)
plt.show()
```
